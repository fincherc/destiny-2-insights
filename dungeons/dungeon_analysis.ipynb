```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the two CSV files into dataframes
df_se = pd.read_csv('../salvations_edge_analysis/se_player_report.csv')
df_vh = pd.read_csv('../vespers_host_analysis/vh_player_report.csv')
```

```python
# Filter data to include only players who successfully cleared the event
df_sd_cleared = df_sd[df_sd['Cleared'] == True]
df_se_cleared = df_se[df_se['Cleared'] == True]
df_vh_cleared = df_vh[df_vh['Cleared'] == True]
```

```python
# Function to plot class distribution
def plot_class_distribution(df, title):
    class_counts = df['Class'].value_counts()
    plt.figure(figsize=(6, 4))
    class_counts.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])
    plt.title(title)
    plt.xlabel('Class')
    plt.ylabel('Number of Players')
    plt.xticks(rotation=0)
    plt.show()
```

```python
# Plot class distribution for SD player report
plot_class_distribution(df_sd_cleared, "Completed Class Distribution in Sundered Doctrine (Top 500 Clears)")
# Plot class distribution for SE player report
plot_class_distribution(df_se_cleared, "Completed Class Distribution in Salvation's Edge (Top 500 Clears)")
# Plot class distribution for VH player report
plot_class_distribution(df_vh_cleared, "Completed Class Distribution in Vesper's Host (Top 500 Clears)")
```

```python
# Compare class distributions between the three datasets
def compare_class_distribution(df1, df2, df3, df1_label='SD Report', 
                               df2_label='VH Report', df3_label='SE Report'):
    class_counts_sd = df1['Class'].value_counts()
    class_counts_vh = df2['Class'].value_counts()
    class_counts_se = df3['Class'].value_counts()

    # Combine the two distributions into a single dataframe for comparison
    comparison_df = pd.DataFrame({
        df1_label: class_counts_sd,
        df2_label: class_counts_vh,
        df3_label: class_counts_se
    }).fillna(0)  # Fill missing values with 0 if a class doesn't exist in one dataset

    # Plot side-by-side comparison of class distributions
    comparison_df.plot(kind='bar', figsize=(8, 6), color=['skyblue', 'lightgreen', 'salmon'])
    plt.title("Completed Class Distribution: Sundered Doctrine vs. Vesper's Host vs. Salvation's Edge (Top 500 Clears)")
    plt.xlabel('Class')
    plt.ylabel('Number of Players')
    plt.xticks(rotation=0)
    plt.show()
```

```python
# Compare class distributions between SE and VH player reports
compare_class_distribution(df_sd_cleared, df_vh_cleared, df_se_cleared)
```
